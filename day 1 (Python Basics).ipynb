```python
# Simple Calculator
def add(a,b):
    return a+b

def sub(a,b):
    return a-b

def mul(a,b):
    return a*b 

def div(a,b):
    try:
        return a/b 
    except Exception as e:
        return e
def SimpleCalculator(num1, num2, option):
    try:
        if option.lower() in ['1', "add", "addition", 'sum']:
            print(add(num1, num2))
        elif option.lower() in ['2', 'sub', 'substraction', 'diff', 'difference']:
            print(sub(num1, num2))
        elif option.lower() in ['3', 'mul', 'multiply', 'product']:
            print(mul(num1, num2))
        elif option.lower() in ['4', 'div', 'division', 'divide']:
            print(div(num1, num2))
        else:
            print("Enter a valid choice.")
    except Exception as e:
        print(e)

print(""" Simple Calculator 
                options : 
                        1. add
                        2. subtract
                        3. multiply
                        4. divide
          Choose a option.""")

option = input("Enter your desired option: ")
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
SimpleCalculator(num1, num2, option)
```

### Report Card for students

```python
def total(marks):
    return sum(marks)

def percentage(n, total):
    return total / n 

def grade(per_of_student):
    if per_of_student >= 90:
        print("Your Grade is: A+")
    elif per_of_student >= 80:
        print("Your Grade is: A")
    elif per_of_student >= 70:
        print("Your Grade is: B")
    elif per_of_student >= 60:
        print("Your Grade is: C")
    else:
        print("Your Grade is: D")

def report_card(n, marks):
    print("""What would you like to see: 
                    1. Total 
                    2. Percentage
                    3. Grade """)
    option = input("Choose the option: ")
    total_marks = total(marks)
    per_of_student = percentage(n, total_marks)
    if option == '1':
        print(f'Total mark of the student is {total_marks} / {n*100}')
    elif option == '2':
        print(f'Percentage: {per_of_student}%')
    elif option == '3':
        grade(per_of_student)

print("Report Card for the Students!!!")
n = int(input("Enter the number of Subjects: "))
marks = []
print("Enter the marks of eack subject one by one.")
for _ in range(n):
    mark = int(input("Enter the mark: "))
    marks.append(mark)
report_card(n, marks)
```
